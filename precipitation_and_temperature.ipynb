```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense


# Yıl ve sıcaklık verilerini içeren alt kümeyi oluştur
grouped = df.groupby('Year').agg({'avg_temp': 'mean'}).reset_index()

# Veriyi ölçeklendirme
scaler = MinMaxScaler()
temperature_data = scaler.fit_transform(grouped['avg_temp'].values.reshape(-1, 1))

# LSTM modeli için veri hazırlama fonksiyonu
def prepare_data(values, n_steps):
    X, y = [], []
    for i in range(len(values) - n_steps):
        X.append(values[i:i + n_steps])
        y.append(values[i + n_steps])
    return np.array(X), np.array(y)

n_steps = 3
X, y = prepare_data(temperature_data, n_steps)

# LSTM modelini oluşturma
def create_model(input_shape):
    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=input_shape))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

# Veriyi LSTM giriş formatına uygun hale getirme
X_train = X.reshape((X.shape[0], X.shape[1], 1))
y_train = y

# Modeli oluşturma ve eğitme
input_shape = (X_train.shape[1], X_train.shape[2])
model = create_model(input_shape)
model.fit(X_train, y_train, epochs=200, verbose=0)

# Gelecek yıllar için tahmin yapma
def forecast(model, input_data, n_steps):
    predictions = []
    input_data = input_data[-n_steps:].reshape((1, n_steps, 1))
    for _ in range(3):  # Gelecek 3 yıl için
        yhat = model.predict(input_data, verbose=0)
        predictions.append(yhat[0, 0])
        input_data = np.append(input_data[:, 1:, :], yhat).reshape((1, n_steps, 1))
    return predictions

future_years = np.arange(2023, 2026)
input_data = temperature_data
scaled_predictions = forecast(model, input_data, n_steps)
predictions = scaler.inverse_transform(np.array(scaled_predictions).reshape(-1, 1)).flatten()

# Tahminleri gösterme
print(f'Future Predictions: {dict(zip(future_years, predictions))}')

# Grafik oluşturma
fig, ax = plt.subplots(figsize=(10, 6))

# Gerçek verileri çizme
ax.plot(grouped['Year'], grouped['avg_temp'], marker='o', linestyle='-', label='Gerçek Sıcaklıklar', color='b')

# Gelecek tahminlerini çizme
ax.plot(future_years, predictions, marker='x', linestyle='--', label='Gelecek Tahminleri', color='r')

# Grafik başlık ve etiketleri
ax.set_title('Yıllık Ortalama Sıcaklık ve Gelecek Tahminleri')
ax.set_xlabel('Yıl')
ax.set_ylabel('Ortalama Sıcaklık')
ax.legend()

# Grid ayarları
ax.grid(True, which='both', axis='both', linestyle=':', linewidth=0.5, color='black', alpha=0.5)

# Sıkıştırma ve sol üst köşeye hizalama
plt.tight_layout()
plt.subplots_adjust(left=0.08, top=0.88)
plt.show()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor


# Bağımlı değişken (production_metric) ve bağımsız değişkenler (avg_temp ve second_feature) belirleyelim
X = df[['avg_temp', 'water_metric']]
y = df['production_metric']

# Veriyi ölçekleyelim
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Veriyi eğitim ve test setlerine bölelim
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Random Forest regresyon modelini tanımlayalım
model = RandomForestRegressor(random_state=42)

# Hiperparametre optimizasyonu için GridSearchCV kullanımı
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]
}
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

# En iyi modeli alalım
best_model = grid_search.best_estimator_

# Test seti üzerinde tahmin yapalım
y_pred = best_model.predict(X_test)

# Gerçek değerler ile tahminleri içeren bir DataFrame oluşturalım
results = pd.DataFrame({'avg_temp': X_test[:, 0], 'water_metric': X_test[:, 1], 'y_test': y_test, 'y_pred': y_pred})

# Scatter plot çizelim
plt.figure(figsize=(10, 6))
plt.scatter(results['avg_temp'], results['y_test'], color='blue', label='Gerçek Değerler')
plt.scatter(results['avg_temp'], results['y_pred'], color='red', marker='x', label='Tahminler')
plt.xlabel('Ortalama Sıcaklık')
plt.ylabel('Üretim Metriği')
plt.title('Random Forest Modeli ile Üretim Metriği Tahmini')
plt.legend()
plt.grid(True)
plt.show()

# Model performansını değerlendirelim
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```
